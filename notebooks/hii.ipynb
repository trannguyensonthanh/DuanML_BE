```python
import os
import zipfile
from pathlib import Path
from tqdm.notebook import tqdm  # Thanh tiáº¿n trÃ¬nh xá»‹n xÃ² cho Jupyter

# --- Cáº¤U HÃŒNH ---
# ÄÆ°á»ng dáº«n Ä‘áº¿n thÆ° má»¥c chá»©a zip (tÆ°Æ¡ng Ä‘á»‘i so vá»›i file notebook náº±m á»Ÿ ngoÃ i)
extract_path = Path("data/processed") 

def extract_all_zips():
    # 1. Kiá»ƒm tra Ä‘Æ°á»ng dáº«n
    if not extract_path.exists():
        print(f"âŒ Lá»—i: KhÃ´ng tÃ¬m tháº¥y thÆ° má»¥c '{extract_path}'")
        return

    # 2. TÃ¬m táº¥t cáº£ file .zip
    zip_files = list(extract_path.glob("*.zip"))
    
    if not zip_files:
        print(f"âš ï¸ KhÃ´ng tÃ¬m tháº¥y file .zip nÃ o trong '{extract_path}'")
        return

    print(f"ğŸ“‚ TÃ¬m tháº¥y {len(zip_files)} file zip. Báº¯t Ä‘áº§u giáº£i nÃ©n vÃ o '{extract_path}'...")

    # 3. VÃ²ng láº·p giáº£i nÃ©n
    for zip_file in tqdm(zip_files, desc="Tiáº¿n Ä‘á»™ giáº£i nÃ©n"):
        try:
            print(f"â¡ï¸ Äang giáº£i nÃ©n: {zip_file.name}...")
            
            with zipfile.ZipFile(zip_file, 'r') as zip_ref:
                # extractall(path) sáº½ giáº£i nÃ©n ná»™i dung vÃ o Ä‘Ãºng thÆ° má»¥c path
                zip_ref.extractall(extract_path)
                
            print(f"âœ… Xong: {zip_file.name}")
            
            # [TÃ™Y CHá»ŒN] XÃ³a file zip sau khi giáº£i nÃ©n xong cho nháº¹ á»• cá»©ng
            # os.remove(zip_file) 
            # print(f"ğŸ—‘ï¸ ÄÃ£ xÃ³a file zip gá»‘c: {zip_file.name}")

        except zipfile.BadZipFile:
            print(f"âŒ Lá»—i: File {zip_file.name} bá»‹ há»ng hoáº·c khÃ´ng pháº£i file zip chuáº©n.")
        except Exception as e:
            print(f"âŒ Lá»—i khÃ´ng xÃ¡c Ä‘á»‹nh vá»›i {zip_file.name}: {e}")

    print("\nğŸ‰ HOÃ€N Táº¤T QUÃ TRÃŒNH GIáº¢I NÃ‰N!")
    print(f"Kiá»ƒm tra láº¡i thÆ° má»¥c: {extract_path.absolute()}")

# Cháº¡y hÃ m
if __name__ == "__main__":
    extract_all_zips()
```
